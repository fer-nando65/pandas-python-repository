```python
import pandas as pd
```

```python
df = pd.read_csv('weather_data_2.csv', parse_dates=['day'])
df
```

```python
df.set_index('day', inplace=True)
df
```

## fillna

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'no event'
})
new_df
```

```python
new_df = df.fillna(method='ffill')
new_df
```

```python
new_df = df.fillna(method='bfill')
new_df
```

```python
new_df = df.fillna(method='bfill', axis='columns')
new_df
```

```python
new_df = df.fillna(method='ffill', limit=1)
new_df
```

## interpolate

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.interpolate(method='time')
new_df
```

## dropna

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how='all')
new_df
```

```python
new_df = df.dropna(thresh=2) #the n minimum value of row can to be save
new_df
```

```python
dt = pd.date_range('01-01-2017','01-11-2017')
idx = pd.DatetimeIndex(dt)
new_df = df.reindex(idx)
new_df
```
